The R-squared score for the test data is 0.990, indicating that the linear regression model explains approximately 99% of the variance in the target variable.

The mean squared error (MSE) for the test data is 3795.53, which represents the average squared difference between the predicted values and the actual values. A lower MSE indicates better model performance.

The root mean squared error (RMSE) for the test data is 61.61. RMSE is the square root of the MSE and represents the average magnitude of the prediction errors. A lower RMSE indicates better model performance.

Overall, these evaluation metrics suggest that the linear regression model performs well on the test data, with high R-squared score and relatively low MSE and RMSE values.
